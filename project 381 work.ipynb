```python
from tkinter import *
import pytesseract
import cv2
import numpy as np
import easyocr

def pictureProcess(): 
    root = Tk()
    root.title("Application")
    root.geometry("730x670")
    root.iconbitmap(r"C:\Users\Admin\Pictures\project icons\recog.ico")
    root.configure(bg='#022D36')

    Label(root,text="Recognition App",bg='#022D36', font = "Helvetica 12 bold italic", fg = "#48AAAD").grid(row=2,column=2)

    Label(root,bg='#022D36').grid()

    Label(root,text="file path",bg='#022D36', font = "Helvetica 12 bold italic", fg = "#7CFC00").grid(row=4,column=1)

    path = Entry(root,width=55)

    path.grid(row=4,column=2)

    Label(root,bg='#022D36').grid()
    Label(root,bg='#022D36').grid()

    def Exit():
        root.destroy()
        
    def processing():
        

        reader = easyocr.Reader(['en'],gpu = False) # load once only in memory.

        image_file_name= path.get()
        image = cv2.imread(image_file_name)

        # sharp the edges or image.
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        sharpen_kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
        sharpen = cv2.filter2D(gray, -1, sharpen_kernel)
        thresh = cv2.threshold(sharpen, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]
        r_easy_ocr=reader.readtext(thresh,detail=0)
        message = ""
        for w in r_easy_ocr:
            message = message + w +" "
        print(text)
        text.insert('1.0',message)
        with open('processText.txt', 'w',encoding="utf-8") as f:
            f.write(str(message))
            f.close()

    def process():
        #img = path.get()
        img = cv2.imread(path.get())

        img = cv2.resize(img, None, fx=1.2, fy=1.2, interpolation=cv2.INTER_CUBIC)
        img = cv2.resize(img, None, fx=2, fy=2)

        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        kernel = np.ones((1,1), np.uint8)
        img = cv2.dilate(img, kernel, iterations=1)
        img = cv2.erode(img, kernel, iterations=1)

        img = cv2.threshold(cv2.medianBlur(img, 3), 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

        cv2.imwrite('thresh.png', img)

        pytesseract.pytesseract.tesseract_cmd = 'C:\\Program Files (x86)\\Tesseract-OCR\\tesseract.exe'

        for psm in range(6,13+1):
            config = '--oem 3 --psm %d' % psm
            txt = pytesseract.image_to_string(img, config = config, lang='eng')
            #print('psm ', psm, ':',txt)
            if psm==13:
                text.insert('1.0',txt)
        everything = text.get('1.0','end')

        with open('processText.txt', 'w',encoding="utf-8") as f:
            f.write(str(everything))


    Button(root,text="Execute",command =processing, relief=RAISED, bg='#D0F0C0', font = "Helvetica 12 bold italic", fg = "black").grid(row=4,column=180)

    text = Text(root,height=30,width=70)
    text.grid(row=7,column=2)

    Label(root,bg='#022D36').grid()

    Button(root,text="Exit",command =Exit, relief=RAISED, bg='#D0F0C0', font = "Helvetica 12 bold italic", fg = "black",width=10).grid(column=2)

    Label(root,bg='#022D36').grid()

    root.mainloop()

```
